```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/kdemertzis/ML_BigData_CyberSecurity-test/master/Ad-Blocking/ad.data", header=None)
df.rename(columns={1558: "label"}, inplace=True)
```

```python
df.head()
```

```python
improper_rows = []
for index, row in df.iterrows():
    for col in df.columns:
        val = str(row[col]).strip()
        if val == "?":
            improper_rows.append(index)
```

```python
df = df.drop(df.index[list(set(improper_rows))])
```

```python
def label_to_numeric(row):
    """Binarize the label."""
    if row["label"] == "ad.":
        return 1
    else:
        return 0
```

```python
df["label"] = df.apply(label_to_numeric, axis=1)
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df)
```

```python
y_train = df_train.pop("label").values
y_test = df_test.pop("label").values
X_train = df_train.values
X_test = df_test.values
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
