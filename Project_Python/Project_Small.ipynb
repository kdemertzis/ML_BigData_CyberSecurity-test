```python
import pandas as pd
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/kdemertzis/ML_BigData_CyberSecurity-test/master/Project_Python/data_small.csv')
```

```python
dataset.head()
```

```python
from pycaret.classification import *
```

```python
classification_setup = classification.setup(data= dataset, target='class', html=False, silent=True, verbose=False)
```

```python
best = compare_models(n_select = 1) 
```

```python
classification.evaluate_model(best)
```
